<span style='font-size:medium'>Part 1: Neutrino Beam Shape Analysis</span>


<span style='font-size:medium'>The first part of the study focuses on determining the shape of the neutrino beam. The muon neutrino oscillation survival probability was plotted as a function of distance for a 2 GeV neutrino, considering parameters such as </span>$\Delta m^{2} = 2.2 \times 10^{-3} eV^{2}$<span style='font-size:medium'> and </span>$\sin^{2}(2\theta) = 0.98$<span style='font-size:medium'>. The Python code for this part includes:</span>



````python
import numpy as np
import matplotlib.pyplot as plt
def P_numu_numu(L, E, deltam2, sin22theta):
    term1 = 1.27 * deltam2 * L / E
    term2 = np.sin(term1) ** 2
    return 1 - term2 * sin22theta

E = 2.0
deltam2 = 2.2e-3
sin22theta = 0.98
Ls = np.linspace(0, 1000, 100)

Ps = P_numu_numu(Ls, E, deltam2, sin22theta)

plt.plot(Ls, Ps)
plt.xlabel('Distance (km)')
plt.ylabel('Survival Probability')
plt.title('Muon Neutrino Oscillation Survival Probability')
plt.grid()
plt.show()
````

1. Import Libraries:
   numpy is imported as np for numerical operations matplotlib.pyplot is imported as plt for plotting.

2. Function Definition \(P\_numu\_numu\): This function calculates the survival probability of muon neutrinos over a distance L.
   Parameters:
   ```
    L: Distance the neutrino travels (in kilometers).

    E: Energy of the neutrino (in GeV).

    deltam2: Difference in the squares of the masses of the neutrino mass eigenstates (in eV²).

    sin22theta: Sine of twice the mixing angle, a parameter that describes the mixing between different neutrino types.
   ```

   The formula used is a simplified version of the neutrino oscillation formula, assuming two\-neutrino mixing. \(Key code segments include the initialization of parameters neutrino energy \(E\), mass difference \(deltam2\), and the mixing angle \(sin22theta\). 

3. Parameter Initialization:
   E, deltam2, and sin22theta are set to specific values. These are constants used in the neutrino oscillation formula.

4. Distance Range \(Ls\):
   Ls is a numpy array of distances ranging from 0 to 1000 km, divided into 100 points. This array represents the different distances at which the survival probability will be calculated.

5. Calculate Survival Probabilities \(Ps\):
   Ps is calculated by applying the P\_numu\_numu function to each distance value in Ls.

6. Plotting:
   The survival probabilities \(Ps\) are plotted against the distances \(Ls\).
   The plot is labeled with an X\-axis \(Distance in km\), Y\-axis \(Survival Probability\), and a title.

7. Display Plot:
   plt.show\(\) displays the plot. 

Overall, this code is used to visualize how the survival probability of muon neutrinos changes as they travel over different distances, demonstrating the concept of neutrino oscillation.

<span style='font-size:medium'>Part 2: Far Detector Spectrum Prediction</span>


In this part, the study advances to predicting the far detector spectrum for different neutrino oscillation parameters. A Gaussian function was employed to model the energy distribution of the neutrinos. The code snippet for this section is:



````python
# gaussian function to fit to get peak energy, width and height
import numpy as np
import scipy

def gaus(x,p0,p1,p2):
    return p0*np.exp((-0.5 * (x - p1)**2) / p2**2)

# fit the ND spectrum
from scipy.optimize import curve_fit

# guess some starting parameters
start_params = (107, 2., 0.45)

# check that the starting parameters are reasonable
plt.hist(bin_edges[:-1], bin_edges, weights = counts, histtype = 'step')
xs = np.arange(0.0, 5.0, 0.001)
plt.plot(xs,gaus(xs, start_params[0], start_params[1], start_params[2]))

plt.xlabel('Energy')
plt.ylabel('Counts')
plt.title('Gaussian Fit')

plt.show
````

This code fits a Gaussian function to a histogram representing an energy spectrum. It involves:

1. Defining a Gaussian Function: A function gaus is defined to model a Gaussian distribution with parameters for height, mean, and standard deviation.
2. Initial Parameter Guess: Initial guesses for these parameters are provided \(start\_params\).
3. Plotting the Histogram: The histogram \(data\) is plotted using plt.hist.
4. Plotting the Initial Gaussian Fit: The Gaussian function with initial parameters is plotted over the histogram using plt.plot.
5. Displaying the Plot: The plot is prepared with labels and a title.



The Gaussian curve, generated using the function gaus with starting parameters start\_params = \(107, 2., 0.45\), is plotted over the histogram. This curve is characterized by its peak \(amplitude\), mean \(central value\), and standard deviation \(width\).

The peak amplitude \(p0 = 107\) suggests the highest point of the Gaussian curve. In the context of this experiment, it might represent the maximum likelihood of neutrino event occurrence at a particular energy level.


The mean \(p1 = 2.0 GeV\) indicates the energy level around which the neutrino events are most densely distributed. This is where the Gaussian curve is centered and can be interpreted as the most probable energy of neutrino events.


The standard deviation \(p2 = 0.45 GeV\) gives an idea of the spread of the energy distribution around the mean. A smaller standard deviation implies a more concentrated distribution of neutrino events around the mean energy, while a larger standard deviation indicates a wider spread.

**Code Snippet:**

	FD_unosc = np.random.poisson(gaus(bin_centers, params[0], params[1], params[2]))

Here, FD\_unosc represents the Poisson fluctuated spectrum for the far detector, and gaus is the Gaussian function with parameters fitted to the near detector data.

<span style='font-size:medium'>Part 3: Fitting Oscillation Parameters</span>



The fitting procedure is crucial to determine the best fit oscillation parameters. The study utilized a curve-fitting algorithm to match the observed data with the predicted model. The fitting code is as follows:



````python
FD_counts = np.array([4, 5, 12, 12, 9, 18, 15, 10, 10, 6, 3, 0, 0, 0, 0, 3, 5, 2, 2, 2, 0, 2])

def FD_func(x, deltam2, sin22theta):
    L = 1000.0
    return gaus(x,params[0], params[1],params[2]) * P_numu_numu(L, x, deltam2, sin22theta)

params = (102.47052429, 2.0004197, 0.38814628)
params_f = (0.00259085, 0.99918752)
FD_fluc = (0, 3, 8, 6, 10, 19, 14, 14, 6, 6, 0, 0, 0, 0, 1, 1, 4, 4, 1, 3, 1, 2)

plt.hist(bin_edges[:-1], bin_edges, weights = FD_counts, histtype = 'step', label = "FD fake data")
plt.hist(bin_edges[:-1], bin_edges, weights = FD_func(bin_centers, params_f[0], params_f[1]), histtype = 'step', label = "FD best fit")
plt.hist(bin_edges[:-1], bin_edges, weights = FD_fluc, histtype = 'step',label = "FD best fit (fluctuated)")
plt.legend()
plt.xlabel('Neutrino Energy (GeV)')
plt.ylabel('Event Rate')
plt.title('Neutrino Energy Spectrum with Oscillations')

plt.show()
````

- **Objective:** The main goal of this part is to fit the far detector neutrino spectrum and determine the best fit values for the oscillation parameters 
  $\sin^{2}2\theta$  and  $\Delta m^{2}$

- **Data and Methodology:** 
  The actual observed far detector counts are provided in the same energy bins as in the near detector, given as FD\_counts.
  The curve\_fit function from scipy.optimize is used for the fitting process. The function to be fit \(FD\_func\) is a combination of a Gaussian function \(representing the neutrino energy distribution\) and the oscillation probability function.
  The x values for fitting are the bin centers in neutrino energy, and the y values are the muon neutrino counts \(FD\_counts\).

- \*\*Code and Calculations: A specific function FD\_func is defined for the fitting, which takes into account the distance to the far detector \(L = 1000.0 km\), the neutrino energy \(x\), and the oscillation parameters. Initial guesses for the fit parameters are provided \(start\_fits\), and curve\_fit is used to find the best fit parameters. 
  A 
$\chi^{2}$
  test is performed to assess the goodness of fit. 
  The uncertainties of the fitted parameters are discussed, with suggestions for further investigation into how changes in experimental conditions might affect the results.

- Results and Visualization: The results include the best\-fit parameters for the neutrino oscillation model, along with their uncertainties.
  A histogram plot is generated to visualize the fit, comparing the observed far detector data, the best fit, and a fluctuated version of the best fit.

- Extension and Further Investigations:
  The section concludes with suggestions for extending the study. These include evaluating how changes in the distance to the far detector, the peak neutrino energy, or the duration of the experiment might impact the results.

This part is crucial in understanding the behavior of neutrinos at the far detector and provides significant insights into the oscillation parameters that govern neutrino behavior. 



````python
start_fits = (2.2e-3, 0.98)
params_f, cov_f = curve_fit(FD_func, bin_centers,FD_counts, start_fits)
print(params_f, cov_f)

def chi2_test(observed,expected):
    return sum(((observed - expected)**2)/expected)
ndof = len(bin_centers)
final_chi2 = chi2_test(FD_counts, FD_func(bin_centers, params_f[0], params_f[1]))
print(final_chi2/ndof)
````

The fitted parameters were found to be $\sin^{2}2\theta = 0.99918752$  and  $\Delta m^{2} = 0.00259085  eV^{2}$, with the chi\-squared value of the fit being $0.7895360925522869$.
